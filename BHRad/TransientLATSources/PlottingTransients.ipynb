```python
from astropy.table import Table
```

```python
tbl = Table.read('./datasets/J0402-1227.fits', hdu=1)
keep_cols = [name for name in tbl.colnames if len(tbl[name].shape) <= 1]
data = tbl[keep_cols].to_pandas()
data
```

```python
seconds_in_30_days = 2.592e6
seconds_in_1_day = 3600*24

print('%.3e'%data['TIME'].min())
print('%.3e'%data['TIME'].max())
```

```python
(data['TIME'].max() - data['TIME'].min())/seconds_in_30_days
```
